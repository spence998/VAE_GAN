```python
import tensorflow as tf
import keras.backend as K
import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
x_train = x_train.astype(np.float32) / 255
x_test = x_test.astype(np.float32) / 255
```

```python
print("x_train shape:", x_train.shape, "y_train shape:", y_train.shape)
print("x_test shape:", x_test.shape, "y_test shape:", y_test.shape)
```

```python
def rounded_accuracy(y_true, y_pred):
    return keras.metrics.binary_accuracy(tf.round(y_true), tf.round(y_pred))
```

```python
class Sampling(keras.layers.Layer):
    def call(self, inputs):
        mean, log_var = inputs
        return K.random_normal(tf.shape(log_var)) * K.exp(log_var / 2) + mean
```

```python
codings_size = 10

inputs = keras.layers.Input(shape=[28,28])
z = keras.layers.Flatten()(inputs)
z = keras.layers.Dense(150, activation="selu")(z)
z = keras.layers.Dense(100, activation="selu")(z)
codings_mean = keras.layers.Dense(codings_size)(z)
codings_log_var = keras.layers.Dense(codings_size)(z)
codings = Sampling()([codings_mean, codings_log_var])
variational_encoder = keras.Model(
    inputs=[inputs],
    outputs=[codings_mean, codings_log_var, codings]
)
```

```python
decoder_inputs = keras.layers.Input(shape=[codings_size])
x = keras.layers.Dense(100, activation="selu")(decoder_inputs)
x = keras.layers.Dense(150, activation="selu")(x)
x = keras.layers.Dense(784, activation="sigmoid")(x)
outputs = keras.layers.Reshape([28,28])(x)
variational_decoder = keras.Model(inputs=[decoder_inputs], outputs=[outputs])
```

```python
_,_,codings = variational_encoder(inputs)
reconstructions = variational_decoder(codings)
variational_ae = keras.Model(inputs=[inputs], outputs=[reconstructions])
```

```python
latent_loss = -0.5 * K.sum(
    1 + codings_log_var - K.exp(codings_log_var) - K.square(codings_mean),
    axis=-1)
variational_ae.add_loss(K.mean(latent_loss) / 784.)
variational_ae.compile(loss="binary_crossentropy", optimizer="rmsprop")
```

```python
history = variational_ae.fit(
    x_train,
    x_train,
    epochs=50,
    batch_size=256,
    validation_data=[x_test, x_test],
)
```

### Generating images

```python
codings = tf.random.normal(shape=[12, codings_size])
images = variational_decoder(codings).numpy()
```

```python
fig, ax = plt.subplots(3, 4)
x = 0
for i in range(3):
    for j in range(4):
        ax[i,j].set_xticks([])
        ax[i,j].set_yticks([])
        ax[i,j].imshow(images[x], cmap="binary")
        x += 1
```

### It is also possible to interpolate the images at the coding level to create an image that is somewhere between two images

```python
codings_grid = tf.reshape(codings, [1,3,4, codings_size])
larger_grid = tf.image.resize(codings_grid, size=[5, 7])
interpolated_codings = tf.reshape(larger_grid, [-1, codings_size])
images = variational_decoder(interpolated_codings).numpy()
```

```python
fig, ax = plt.subplots(5, 7)
x = 0
for i in range(5):
    for j in range(7):
        if i % 2 == 0 and j % 2 == 0:
            plt.setp(ax[i,j].spines.values(), color="red")
        ax[i,j].set_xticks([])
        ax[i,j].set_yticks([])
        ax[i,j].imshow(images[x], cmap="binary")
        x += 1
```
